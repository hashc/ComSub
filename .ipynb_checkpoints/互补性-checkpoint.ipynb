# 互补性

## 商品的互补和替代关系：

1.互补：奶嘴和纸尿裤。
2.替代：有A就没B，有B就没A，比如纸尿裤和布尿裤。
纸尿裤和布尿裤，都和婴儿用品距离很近，但他们不会被同时购买，所以他们构成了替代关系（竞争）。在技能上可以定义类似的关系，比如Python和R也是可替代的。

为什么会出现可替代的关系？

商品需要金钱购买，技能、技能需要时间去学习，当仅有有限的资源，表现在共现矩阵上，就是大多数人不会同时拥有竞争类的产品和技能

因此，对于可替代的商品，要求他们在全局上都是相似的，但是共现矩阵上确是共同出现概率很小。

对于互补的商品，要求他们在全局上是不相似，但在共现矩阵上共同出现的概率很大。


 -|共现概率低|共现概率高|
-|-|-|
结构不相似|？|互补（结构洞上的桥？）
结构相似|可替代|互补



```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import scipy.sparse.linalg 
import scipy
from scipy import sparse
import numpy as np
import matplotlib.pyplot as plt
import sys
import seaborn as sns
from ECI import *
from collections import defaultdict
from numpy import linalg as la
```

```python
def true_ivide(a,b):
    # a/b ,解决b中含有0元素或者含有nan的问题
    with np.errstate(divide='ignore', invalid='ignore'):
        c = np.true_divide(a,b)
        c[c == np.inf] = 0
        c = np.nan_to_num(c)
    return c
def get_prab(M):
    # 按照行归一化M M(i,j)表示 P(j|i) 
    return true_ivide(M,M.sum(1)[:, np.newaxis])
def stats_bins(X,Y,bins = np.linspace(0,1,11)):
    bins=np.linspace(0,1,11)
    xm = max(bins)
    XY = [(x,y) if x<=xm else (xm,y) for (x,y) in zip(X,Y)]
    XY = sorted(XY,key=lambda x:x[0])
    n = len(bins)
    yb = [[] for i in range(n-1)]
    s=0
    for (x,y) in XY:
        if x<=bins[s+1]:
            yb[s].append(y)
        else:
            s+=1
            yb[s].append(y)
    x1 = bins[1:]
    return yb
def compare_Glo(B,PM,index = [(i,j) for i in range(nR) for j in range(nR) if PM[i][j] and i>j]):
    nR= len(Remain)
    X,Y=[],[]
    for (i,j) in index:
        X.append(B[i][i])
        Y.append(PM[i][j])
    return X,Y
```

```python
percentiles = [25,50,75,90]
cuts=np.percentile(wage,percentiles) #收入的前25%,前50%,前75%，前90%
print(percentiles)
print(cuts)
```

```python
import copy
tmpDict = {'P':np.zeros((len(NameDict),len(NameDict))),'M':np.zeros((len(NameDict),len(NameDict)))}
TotalData={}
for key in [0]+percentiles:
    TotalData[key] = copy.deepcopy(tmpDict)
for tmp,w in zip(skills_new,wage):
    for sk0 in tmp:
        for sk1 in tmp:
            for key,cut in zip([0]+percentiles,[0]+list(cuts)):
                if w>=cut:
                    TotalData[key]['M'][NameDict[sk0]][NameDict[sk1]]+=1
                    TotalData[key]['M'][NameDict[sk1]][NameDict[sk0]]+=1
                    TotalData[key]['P'][NameDict[sk0]][NameDict[sk1]]+=w
                    TotalData[key]['P'][NameDict[sk1]][NameDict[sk0]]+=w
                   
```

```python
TotalPMMat={}
for cut in [0]+percentiles:
    TotalPMMat[cut]=true_ivide(TotalData[cut]['P'],TotalData[cut]['M'])
```
